```python
import numpy as np
import matplotlib.pyplot as plt
```

## Neural network using activation function the Sigmoid : f(x) = 1/(1+ex) 

```python
def g(x):
    if 0 <= x <= 0.5:
        return 1/4 + x
    elif 0.5 < x <= 1:
        return 5/4 + x
    else:
        return 0
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

## Use Relu activation function to do the approximation

```python
def relu(x):
    if x<0:
        return 0
    else:
        return x
```

## Plot the functions and approximations

```python
# Create data for plotting
x_values = np.linspace(-1, 2, 1000)
y_values_g = [g(x) for x in x_values]
```

```python
y_values_sigmoid = [sigmoid(x) for x in x_values]
```

```python
y_values_relu = [relu(x) for x in x_values]
```

```python
plt.plot(x_values, y_values_g, label='g(x)', color='black')
```

```python
plt.plot(x_values, y_values_sigmoid, label='Sigmoid Approximation', linestyle='--')
```

```python
plt.plot(x_values, y_values_relu, label='ReLU Approximation', linestyle='--')
```
